```python
os.chdir("playlists")

html_files = [file for file in glob.glob("*.html")]

html_files
```

```python
for file in html_files:
    print(os.path.join("playlists", file))
```

```python
df_all = pd.DataFrame(columns=['filename', 'sevenDigitalIdentifier', 'artist', 'title', 'position'])

for file in html_files:
    with open(os.path.join("../playlists", file)) as f:
        soup = BeautifulSoup(f.read(), 'html.parser')
    
    italy = soup.find('li', id="bulk_tracks-tab-IT")
    all_textarea = italy.find_all('textarea')
    
    file_list = []
    sdi = []
    artist_list = []
    title_list = []
    position = range(1,len(all_textarea)+1)

    for text in all_textarea:
        test_dict = json.loads(text.text)
        file_list.append(file)
        sdi.append(test_dict["sevenDigitalIdentifier"])
        artist_list.append(test_dict["artist"])
        title_list.append(test_dict["title"])

    df = pd.DataFrame(columns=['sevenDigitalIdentifier', 'artist', 'title', 'position'])
    df['filename']=file_list
    df['sevenDigitalIdentifier']=sdi
    df['artist']=artist_list
    df['title']=title_list
    df['position']=position
    
    df_all = pd.concat([df_all,df], ignore_index=True)
```
