```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from pandas.io import gbq

pd.set_option('display.float_format', lambda x: '%.3f' % x)
%matplotlib inline
```

```python
streams_sql = "SELECT * FROM `umg-comm-tech-dev.adhoc.sav_streams`"
data = gbq.read_gbq(query=streams_sql, project_id='umg-comm-tech-dev', dialect='standard')
data.head()
```

```python
data.shape
```

```python
data = data.drop_duplicates()
data.shape
```

```python
data['date'] = pd.to_datetime(data[['year', 'month']].assign(day=1))
```

```python
data.head()
```

```python
by_uri_date = data.groupby(by=['source_uri','date']).monthly_streams.sum().reset_index()
by_uri_date.sort_values(by=['source_uri','date'], ascending=[True,True])
by_uri_date.head()
```

```python
uri_map = pd.read_clipboard()
uri_map.head()
```

```python
by_uri_date = by_uri_date.merge(uri_map, how='left', on='source_uri')
by_uri_date.head()
```

```python
sns.relplot(x="date", y="monthly_streams",
            hue="name",
            height=5, aspect=1.75, facet_kws=dict(sharex=False),
            kind="line", legend="full", data=by_uri_date)
```

```python
by_uri_date.groupby(by=['name','date']).monthly_streams.sum()
```

```python
data.to_csv('sav_isrc_level.csv')
```

```python
by_uri_date.to_csv('sav_playlist_level.csv')
```

```python
streams_sql = "SELECT * FROM `umg-comm-tech-dev.adhoc.streams_sav_2`"
data = gbq.read_gbq(query=streams_sql, project_id='umg-comm-tech-dev', dialect='standard')
data.head()
```

```python
data.shape
```

```python
data.describe(include='all')
```

```python
data['upc2'] = ['upc_'+str(x) for x in data.upc]
```

```python
data.upc.unique()
```

```python
data.streams.sum()
```

```python
g = sns.relplot(x="stream_date", y="streams",
            hue="upc2",
            height=5, aspect=1.75, facet_kws=dict(sharex=False),
            kind="line", legend="full", data=data)
g.fig.get_axes()[0].set_yscale('log')
plt.show()
```

```python
data['alt_xmas'] = [1 if x=='00600406837608' else 0 for x in data.upc]
```

```python
data['xmas_date'] = [1 if x.month in [11,12] else 0 for x in data.stream_date]
data['year'] = [x.year for x in data.stream_date]
```

```python
xmas_dates = data[data.xmas_date==1]
```
