```python
import pandas as pd
from pandas.io import gbq
```

```python
playlists1 = pd.read_csv('/Users/daria/Downloads/compilations_prod_2019-03-26.csv')
playlists1.head()
```

```python
playlists1 = playlists1[playlists1.UPC.isnull()==False]
```

```python
playlists1['UPC'] = ['00'+str(x)[:-2] for x in playlists1.UPC]
playlists1.UPC
```

```python
playlists1.head()
```

```python
playlists2 = pd.read_csv('/Users/daria/Downloads/Fixed Playlist Release Schedule - Release Schedule.csv')
playlists2.head()
```

```python
playlists2 = playlists2.rename(columns={'Title/Theme':'Release title'})
```

```python
playlists2 = playlists2.drop(columns=['Status','List Formatted UPC','Spotify-Only UPC (If Applicable)'])
```

```python
playlists2.head()
```

```python
playlists = pd.concat([playlists1,playlists2], ignore_index=True)
playlists.head()
```

```python
playlists.UPC = [str(x) for x in playlists.UPC]
```

```python
playlists['Release Date'] = pd.to_datetime(playlists['Release Date'])
```

```python
playlists.UPC.unique()
```

```python
playlists_cols = [str(x).lower().replace(' ','_') for x in playlists.columns]
playlists_cols
```

```python
playlists.columns = playlists_cols
playlists.head()
```

```python
playlists.to_gbq('fixed_playlists_data.playlists_list', project_id='umg-comm-tech-dev', if_exists='replace')
```
