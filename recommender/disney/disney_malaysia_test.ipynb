```python
import pandas as pd
import pickle

import time

from pitched_recommend import Recommender
```

```python
with open('/Users/daria/Downloads/recommender%2Fmodel_20181112.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
data = pd.read_csv('/Users/daria/Downloads/results-20181129-120148.csv')
data.head()
```

```python
data2 = pd.read_csv('/Users/daria/Downloads/results-20181129-164518.csv')
data2 = data2.rename(columns={'tm_isrc':'dcr_isrc'})
data_all = pd.concat([data,data2], ignore_index=True)

data_all.head()
```

```python
for_model = data_all[data_all.repertoire_owner=='Walt Disney Records']
for_model = for_model.reset_index(drop=True)
```

```python
top20 = for_model.sort_values(by=['streams'], ascending=False).reset_index().iloc[:20,:]
top20.head()
```

```python
cols = ['isrc','track_artist','track_title','days_on_playlist','playlist_uri']
top20['days_on_playlist']=50
top20['playlist_uri']='malaysia_top_test'
top20['isrc']=top20.dcr_isrc

top20 = top20[cols]
```

```python
rec = model.recommend_outofmodel(top20, model.playlists_trainset, N=300)
rec_malaysia = model.rec_readable(rec)
```
