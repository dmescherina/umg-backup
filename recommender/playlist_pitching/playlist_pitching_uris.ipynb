```python
from pandas.io import gbq
import pandas as pd
```

```python
df = gbq.read_gbq('SELECT * FROM `umg-comm-tech-dev.adhoc.browse_pl_extract`', project_id='umg-comm-tech-dev', 
                  dialect='standard')
```

```python
df.head()
```

```python
df = df[df.owner_id=='spotify']
df.head()
```

```python
pl_count = df.groupby(by='playlist_uri')
pl_count_df = pd.DataFrame(pl_count.playlist_name.count()).reset_index()
pl_count_df.head()
```

```python
df = df.reset_index()
df = df.merge(pl_count_df, how='left', left_on='playlist_uri', right_on = 'playlist_uri')
df.head()
```

```python
df = df.rename(columns={'playlist_name_y':'count_appearance'})
df.head()
```

```python
pl_streams = gbq.read_gbq('SELECT * FROM `umg-comm-tech-dev.adhoc.christine_playlist_listen_counts`', 
                          project_id='umg-comm-tech-dev', dialect='standard')
pl_streams.head()
```

```python
pl_streams = pl_streams.drop(columns=['ranking','pl_ranking'])
```

```python
pl_streams['playlist_uri'] = ['spotify:user:'+str(x) for x in pl_streams.playlist_uri]
pl_streams.head()
```

```python
df = df.merge(pl_streams, how = 'left', left_on = 'playlist_uri', right_on = 'playlist_uri')
df.head()
```

```python
df.describe(include='all')
```

```python
df['same_as_country'] = [1 if x==y else 0 for (x,y) in zip(df.territory, df.country_owner)]
```

```python
df['local'] = [1 if x < 20 else 0 for x in df.count_appearance]
```

```python
terr_cat_mean_listeners = df.groupby(by=['territory','category_id']).listeners.mean()
```

```python
terr_cat_mean_listeners = terr_cat_mean_listeners.reset_index()
```

```python
df = df.merge(terr_cat_mean_listeners, how='left', left_on = ['territory','category_id'], 
              right_on=['territory','category_id'])
df.head()
```

```python
df = df.rename(columns={'listeners_x':'listeners','listeners_y':'mean_listeners_cat_terr'})
df = df.drop(columns='playlist_name')
df.head()
```

```python
df['popular'] = [1 if x>y else 0 for (x,y) in zip(df.listeners,df.mean_listeners_cat_terr)]
df.popular.describe()
```

```python
df['score'] = df['same_as_country']+df['local']+df['popular']
```

```python
df = df.sort_values(by=['territory','category_id','score'], ascending=False)
```

```python
df.shape
```

```python
by_test = df[df.score>0].groupby(by=['territory','category_id']).score.sum()
```

```python
by_test_df = pd.DataFrame(by_test).reset_index()
```

```python
by_test_df[by_test_df.score==0]
```

```python
df[df.score>0].to_csv('playlists_for_browse_cat.csv')
```
